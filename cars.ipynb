```python
# drops unnecessary columns
cars = cars.drop(['name', 'year'], axis=1)
cars.head(4)
```

```python
cars = pd.get_dummies(cars, columns=['fuel', 'seller_type', 'transmission'], drop_first=True, dtype=int)
cars.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder

# Define the desired order
owner_order = [['Fourth & Above Owner', 'Third Owner', 'Second Owner', 'First Owner', 'Test Drive Car']]

# Create encoder
oe = OrdinalEncoder(categories=owner_order)

# Apply to the DataFrame
cars[['owner']] = oe.fit_transform(cars[['owner']])

```

```python
## For your Understanding ## Why We passing [['owner']]

# Using a Series (single bracket)
owner_series = cars['owner']
print(owner_series.shape)  # Output will be (n,) - 1D

# Using a DataFrame (double bracket)
owner_dataframe = cars[['owner']]
print(owner_dataframe.shape)  # Output will be (n, 1) - 2D
```

```python
# Step 3: Split the data into training and testing sets (80% train, 20% test)
cars.head()
x = cars.drop('selling_price', axis=1)
y = cars['selling_price']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape )
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
#Make predictions on the test set
y_pred = model.predict(X_test)
```

```python
print(f"{y_pred},\n{y_test}")
```

```python
# Evaluate using R-squared score and Mean Squared Error (MSE)
from sklearn import metrics
r2_score = model.score(X_test, y_test)
mse = metrics.root_mean_squared_error(y_test, y_pred)

print(f'R-squared: {r2_score}')
print(f'Root Mean Squared Error: {mse}')
```
